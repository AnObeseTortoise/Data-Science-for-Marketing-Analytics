```python
import pandas as pd
```

```python
ads = pd.read_csv('Activity2.01/Advertising.csv')
ads.head()
```

```python
ads.info()
```

```python
#Question 1: What are the unique values present in the "Products" Column?

ads["Products"].unique()
```

```python
#Q2: How many data points belong to each category in the "Products" Column?

ads["Products"].value_counts()
```

```python
#Q3: What are the total views across each category in the "Products" Column?

ads.groupby('Products')[['Web','Newspaper','Radio','TV']].sum()

```

```python
#Q4: Which product has the highest viewership on TV?

ads.groupby('Products').sum().plot(kind = 'bar', y = "TV")

#Answer: Mobiles
```

```python
#Q5: Which Product has the lowest viewership on the web?

ads.groupby('Products').sum().plot(kind = 'bar', y = "Web")

#Answer: Laptops
```

```python
#Plot the relationship between the variables in the datasets with the helpof pairplots

import seaborn as sns

sns.pairplot(ads, hue = "Products" )
```
