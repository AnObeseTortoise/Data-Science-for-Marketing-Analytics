```python
"""Clean the DataFrame and make it into Usable form."""


import pandas as pd
```

```python
#import the dataset

sales = pd.read_csv("Activity1.01/sales.csv")

sales.head()
```

```python
#analyze the data Type of the fields

sales.dtypes
```

```python
#adjust the headers in sales dataset 
sales ['Product line'] = sales['Product']+ ' '+sales['line']

sales = sales.drop(['Product' , 'line'], axis = 1)

sales.head()
```

```python
sales['Product'] = sales['Product.2']+ ' '+sales['Order']+ ' '+sales['method']

sales = sales.drop(['Product.2', 'Order','method'], axis = 1) 

sales.head()
```

```python
sales['Order Method'] = sales['type.1']
sales = sales.drop(['type.1'], axis = 1)

sales.head()
```

```python
sales['Product Type'] = sales['Product.1']+ ' '+sales['type']
sales = sales.drop(['Product.1', 'type'], axis = 1)

sales.head()
```

```python
sales = sales.fillna('')
```

```python
sales['Retailer Country'] = sales['Retailer']+' '+sales['country']
sales = sales.drop(['Retailer','country'], axis = 1)

sales.head()
```
